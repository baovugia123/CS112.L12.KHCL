**ĐỊA LAN**

Giáo sư Braun đã lai giống tạo ra ***n*** loại địa lan mới độc đáo. Mỗi loại địa lan được trồng trong một chậu riêng và được ghi một số thể hiện đặc điểm gene của loại hoa đó, chậu thứ i có số là ***ai, i*** = 1 ÷ ***n***.

Hội chợ hoa địa lan thế giới mời giáo sư tham gia và giới thiệu các thành quả nghiên cứu của ông. Các chậu hoa do ông mang tới sẽ được đặt trên k đôn dọc theo đường đi dẫn tới gian hàng chính của hội chợ. 

Muốn gây ấn tượng mạnh cho những người tham dự, đặc biệt là các nhà khoa học Giáo sư yêu cầu trợ lý của mình chọn ***k*** chậu sao cho khi thực hiện phép AND số trên các chậu hoa tới (nếu xét chúng ở dạng biểu diễn nhị phân) sẽ có kết quả bằng 0.

Hãy xác định xem người trợ lý có thể chọn được ***k*** chậu như vậy hay không và đưa ra thông báo ***“YES”*** hoặc ***“NO”***.

***Dữ liệu:*** Vào từ thiết bị nhập chuẩn:
* Dòng đầu tiên chứa 2 số nguyên n và k (1 ≤ ***n*** ≤ 2×10^4, 1 ≤ ***k*** ≤ ***n***),
* Dòng thứ 2 chứa ***n*** số nguyên ***a1, a2, . . ., an*** (0 ≤ ***ai*** < 2^12, ***i*** = 1 ÷ ***n***).

***Kết quả:*** Đưa ra thiết bị xuất chuẩn thông báo tìm được.

**Abstraction:**
- Chọn $k$ số liên tiếp trên dãy số nguyên $a$ gồm n phần tử đã cho trước sao cho AND của chúng bằng 0.
- Thuật toán tham lam.

**Pattern Recognition:**
- Với $k \geq 12$, ta dễ dàng thấy được kết quả sẽ là YES nếu như AND của cả dãy $a$ bằng 0 vì mỗi phần tử dãy $a$ nhiều nhất có 12 bit (xét yêu cầu đề trên 12 bit vì $a_i \le 2^{12}, i = 1 ÷ n$) mà AND của cả dãy tồn bằng 0 chứng tỏ cả 12 bit đó đều có số bằng 0. Vì vậy ta chỉ cần chọn 12 số đó là được.
- Bài toán quy về $k < 12$. Ta sẽ duyệt trâu hết $k$ phần tử liên tiếp trên dãy $a$ với mỗi vị trí bắt đầu $i$ tương ứng.

**Algorithm designed:**
* *Input:*
    - n, k là số nguyên, (1 ≤ ***n*** ≤ 2×10^4, 1 ≤ ***k*** ≤ ***n***). 
    - a là dãy số nguyên, ***a1, a2, . . ., an*** (0 ≤ ***ai*** < 2^12, ***i*** = 1 ÷ ***n***). 
* *Output:* Đưa ra thiết bị xuất chuẩn thông báo tìm được: *YES* hoặc *NO*
* *Mã giả:*
```
Nhập n, k và dãy a
Nếu k >= 12:
  Nếu AND cả dãy a == 0:
    return "YES"
else :
  for i in range(n - k + 1):
    sum = a_i
    for j in range(k):
      sum = sum AND a_{i + j}
    Nếu sum == 0:
      return "YES"
return "NO"
``` 
*Độ phức tạp thời gian: T = $O(nlogn)$

````
n, k = map(int, input().strip().split())
a = list(map(int, input().strip().split()))

if (k >= 12):
    temp = a[0]
    for i in a:
        temp = (temp & i)
    if (temp==0):
        print("YES")
    else:
        print("NO")
    
else:
    flag = True
    for i in range(n - k + 1):
        if (flag == False):
            break
        temp = a[i]
        for j in range(k):
            temp = (temp & a[i + j])
        if (temp == 0):
            flag = False
            break
    if (flag == False):
          print("YES")
    else:
          print("NO")
````
